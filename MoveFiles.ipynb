```python
import shutil
import pandas as pd
import os
```

```python
df = pd.read_csv("/home/ivana/data/all_data_info.csv")
df = df.loc[(df['in_train'] == True) & df['new_filename'].str.startswith('2')]
print(df.shape)
df.head()
```

```python
num_train = df.shape[0] * 80 // 100
num_val = (df.shape[0] - num_train) // 2
num_test = num_val
num_samples = num_train + num_val + num_test

temp = df['artist'].value_counts()
artists = temp[temp >= 0].index.tolist()
print('Kreirat ću ' + str(len(artists)) + ' podmapa')
print('Prebacit ču ' + str(num_samples) + ' slika')
```

```python
# for a in artists:
trainImages = df[:num_train]
valImages = df[num_train:num_train+num_val]
testImages = df[num_train+num_val:]

print(trainImages.shape)
print(valImages.shape)
print(testImages.shape)
```

```python
for index, row in trainImages.iterrows():
    directory = 'data/images/train/' + row['artist']
    
    if not os.path.exists(directory):
        os.makedirs(directory)
        print('Napravila mapu ' + directory)
    
    shutil.move('data/train_2/' + row['new_filename'], directory)
    # print('Prebacila sliku ' + row['new_filename'])
```

```python
for index, row in valImages.iterrows():
    directory = 'data/images/validation/' + row['artist']
    
    if not os.path.exists(directory):
        os.makedirs(directory)
        print('Napravila mapu ' + directory)
    
    shutil.move('data/train_2/' + row['new_filename'], directory)
    print('Prebacila sliku ' + row['new_filename'])
```

```python
for index, row in testImages.iterrows():
    shutil.move('data/train_2/' + row['new_filename'], 'data/images/test/')
    print('Prebacila sliku ' + row['new_filename'])
```
